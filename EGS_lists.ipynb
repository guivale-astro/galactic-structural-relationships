```python
df_z_mass = df_egs_mass.loc[(df_egs_reds["z_best"] < 2.5) & (df_egs_reds["z_best"] > 1.5)]
df_z_phys = df_egs_phys.loc[(df_egs_reds["z_best"] < 2.5) & (df_egs_reds["z_best"] > 1.5)]
df_z_reds = df_egs_reds.loc[(df_egs_reds["z_best"] < 2.5) & (df_egs_reds["z_best"] > 1.5)]
df_z_mult = df_egs_mult.loc[(df_egs_reds["z_best"] < 2.5) & (df_egs_reds["z_best"] > 1.5)]

display(df_z_mass)
print()
display(df_z_phys)
print()
display(df_z_reds)
print()
display(df_z_mult)
print()

print('n_galaxias =', df_z_reds.shape[0])
```

```python
redshift = df_egs_reds[["z_best"]]
redshift.columns = ['Redshift']

eleccion = 1

if eleccion == 1:
    masa = df_egs_mass[['M_neb_med']]
    masa.columns = ['Masa']
elif eleccion == 2:
    masa = df_egs_phys[['M_l99_11a_tau']]
    masa.columns = ['log(M/M_Sun)']

mag_abs = df_egs_phys[['VMag_6a_tau^NEB']]
mag_abs = mag_abs[mag_abs > -90]
mag_abs.columns = ['Magnitud absoluta V']

color_BV = pd.DataFrame(df_egs_phys['BMag_6a_tau^NEB'] - df_egs_phys['VMag_6a_tau^NEB'])
color_BV = color_BV[color_BV > -90]
color_BV.columns = ['Índice de color B-V']

#exc_BV = df_phys['EBV_14a']
#exc_BV.columns = ['E(B-V)']

SFR = np.log10(df_egs_phys[['SFR_14a']])
SFR.columns = ['log(SFR)']
#isoarea =1 

isoarea = np.log10(df_egs_mult[['ISOAREAF_IMAGE']])
isoarea.columns = ['log(isoarea)']

df_test = pd.concat([redshift, masa, mag_abs, color_BV, SFR, isoarea], axis = 1)

sns.set_style("darkgrid")

plt.figure(figsize = (30, 30))
import matplotlib
matplotlib.rc('xtick', labelsize=10) 
matplotlib.rc('ytick', labelsize=10) 


pp = sns.PairGrid(
    df_test,
    #plot_kws=dict(marker="o", linewidth=1, s = 1, edgecolor = 'none'),
    #diag_kws=dict(fill=True),
    #corner = False,
)


# def axes_label(x, **kws):
    #ax = plt.gca()
    #ax.text(
    #    .55, .5, x.name, size=14,
    #    ha="center", va="center", transform=ax.transAxes
    #)



#pp.axes[4, 4].set(xlim = (-0.05, None), ylim = (-0.05, None))


pp.map_lower(sns.kdeplot, fill=False, color = 'gray')
pp.map_diag(sns.histplot, kde=True, color = 'coral', bins = 15, alpha = 0.8)
#pp.map_diag(axes_label)
pp.map_upper(sns.scatterplot, data = df_test, marker = "o", s = 3, edgecolor = 'none', hue = 'Índice de color B-V', palette = 'plasma', alpha = 1)
pp.add_legend(title = 'Índice de color (B-V)')
#plt.show()
matplotlib.rcdefaults()
```

```python
plt.figure(figsize = (12, 10))

df_egs_phys = df_egs_phys[df_egs_phys > -75]

sns.set_style('whitegrid')
matplotlib.rc('xtick', labelsize=25) 
matplotlib.rc('ytick', labelsize=25) 
sns.set_context('talk')

color_UR = pd.DataFrame(df_egs_phys['BMag_6a_tau^NEB'] - df_egs_phys['VMag_6a_tau^NEB'])
color_UR.columns = ['U-R']

mag_R = pd.DataFrame(df_egs_phys['VMag_6a_tau^NEB'])
mag_R.columns = ['Mag_R']

color_BV = pd.DataFrame(df_egs_phys['BMag_6a_tau^NEB'] - df_egs_phys['VMag_6a_tau^NEB'])
color_BV.columns = ['B-V']

sp = sns.scatterplot(x = mag_R['Mag_R'], y = color_UR['U-R'], s = 30, edgecolor = 'none', data = color_BV, hue = 'B-V', palette = 'rainbow', alpha = 0.7)
kdep = sns.kdeplot(x = mag_R['Mag_R'], y = color_UR['U-R'], color = '#2b2b2b', linewidths = 2)#, s = 10, edgecolor = 'none', data = color_BV, hue = 'B-V', palette = 'plasma')
#sp.colorbar(sp)
plt.xlabel('$\mathregular{M_V}$', size = 30)
plt.ylabel('$\mathregular{B-V}$', size = 30)
plt.gca().invert_xaxis()

matplotlib.rcdefaults()
```
