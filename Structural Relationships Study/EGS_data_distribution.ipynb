```python
import pandas as pd
import pandas.plotting as pdplt
import numpy as np
import astropy as ast
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
```

```python
df_mass = pd.read_csv("EGS_stellar_mass(z-GVA).csv")
df_mass = df_mass[df_mass > -90]

df_phys = pd.read_csv("EGS_physical_properties(z-GVA).csv")
df_phys = df_phys[df_phys > -90]

df_reds = pd.read_csv("EGS_redshift_(z-GVA).csv")

df_mult = pd.read_csv("EGS_multiband_(z-GVA).csv")
```

```python
plt.figure(figsize = (15, 8))

#plt.grid()
#bins = 20
import matplotlib
matplotlib.rcParams['axes.axisbelow'] = False

sns.set_style("whitegrid")
sns.set_context('talk')
bins = 1 + np.log2(len(df_phys))
bins = int(bins.round())
sns.histplot(df_reds["z_best"], bins = bins, color = "coral", alpha = 0.7, edgecolor = 'black', linewidth = 2)
#sns.kdeplot(df_reds["z_best"], fill = True, linewidth = 20)

min_ylim, max_ylim = plt.ylim()
x = df_reds["z_best"]
plt.axvline(x.mean(), color='k', linestyle='dashed', linewidth=3)
plt.text(x.mean()*1.01, max_ylim*0.9, 'Valor medio: {:.2f}'.format(x.mean()), fontsize = 30)

matplotlib.rc('xtick', labelsize=25) 
matplotlib.rc('ytick', labelsize=25) 

plt.xlabel('$\mathregular{z_{best}}$', fontsize = 30)
plt.ylabel('Nmero de ocurrencias', fontsize = 30)
plt.ylabel(' ', fontsize = 1)

plt.xticks(np.round(np.arange(1.535, 2.52, step= (2.5-1.51)/15), decimals=2))
plt.title('Distribucin de redshift en 1.5  z  2.5', fontsize = 35)

#plt.figtext(0.5, 0.01, "Para el nmero de bins se ha empleado: $\mathregular{bins = 1+ceil(log_2(n}$ $\mathregular{datos))}$ ", ha="center", fontsize=30)

plt.plot()

matplotlib.rcdefaults()
```
